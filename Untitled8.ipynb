```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

# Generate random data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)  # Normally distributed data

# Plot the data's histogram and overlay the KDE
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

# Histogram of the data
sns.histplot(data, bins=30, kde=False, color='blue', alpha=0.6, label='Histogram')

# KDE using Seaborn
sns.kdeplot(data, color='red', linewidth=2, label='KDE (Seaborn)')

# KDE using Scikit-learn
kde_skl = KernelDensity(kernel='gaussian', bandwidth=0.3)
kde_skl.fit(data[:, None])  # Fit requires a 2D array
x_vals = np.linspace(-4, 4, 1000)[:, None]  # Generate points for evaluation
log_density = kde_skl.score_samples(x_vals)  # Log density values
plt.plot(x_vals, np.exp(log_density), color='green', linestyle='--', label='KDE (Sklearn)')

plt.title('Kernel Density Estimation (KDE)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Penguins dataset
penguins = sns.load_dataset("penguins")

# Drop rows with missing values
penguins = penguins.dropna()

# Plot the bivariate distribution
plt.figure(figsize=(10, 6))
kde = sns.kdeplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    cmap="Blues",
    fill=True,
    thresh=0.05
)

# Add colorbar
kde.collections[0].colorbar = plt.colorbar(kde.collections[0], label="Density")

plt.title('Bivariate Distribution of Bill Length vs. Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')

plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import GeoJSONDataSource, ColumnDataSource, HoverTool
from bokeh.io import output_notebook
import json

# Enable notebook output (optional)
output_notebook()

# Create synthetic geospatial data (GeoJSON format)
geojson_data = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "properties": {"country": "Country A", "population": 5000000},
            "geometry": {
                "type": "Polygon",
                "coordinates": [
                    [
                        [-10, 40],
                        [-10, 50],
                        [0, 50],
                        [0, 40],
                        [-10, 40]
                    ]
                ]
            },
        },
        {
            "type": "Feature",
            "properties": {"country": "Country B", "population": 8000000},
            "geometry": {
                "type": "Polygon",
                "coordinates": [
                    [
                        [5, 45],
                        [5, 55],
                        [15, 55],
                        [15, 45],
                        [5, 45]
                    ]
                ]
            },
        },
    ]
}

# Create city point data
cities = {
    "longitude": [-5, 10],
    "latitude": [45, 50],
    "city": ["City X", "City Y"],
    "population": [200000, 300000],
}

# Create GeoJSONDataSource for polygons
geo_source = GeoJSONDataSource(geojson=json.dumps(geojson_data))

# Create ColumnDataSource for city points
city_source = ColumnDataSource(cities)

# Create Bokeh figure
plot = figure(
    title="Geospatial Data Visualization",
    x_axis_label="Longitude",
    y_axis_label="Latitude",
    tools="pan, wheel_zoom, reset, hover",
    width=800,
    height=600,
)

# Plot polygons (countries)
plot.patches(
    "xs",
    "ys",
    source=geo_source,
    fill_color="lightblue",
    line_color="black",
    line_width=1,
    fill_alpha=0.7,
)

# Plot points (cities)
plot.circle(
    x="longitude",
    y="latitude",
    size=10,
    fill_color="red",
    line_color="black",
    source=city_source,
    legend_label="Cities",
)

# Add hover tool for polygons
hover_polygons = HoverTool(
    tooltips=[("Country", "@country"), ("Population", "@population")],
    renderers=[plot.renderers[0]],
)
plot.add_tools(hover_polygons)

# Add hover tool for city points
hover_cities = HoverTool(
    tooltips=[("City", "@city"), ("Population", "@population")],
    renderers=[plot.renderers[1]],
)
plot.add_tools(hover_cities)

# Add legend
plot.legend.location = "top_left"
plot.legend.title = "Legend"

# Show the interactive plot
show(plot)
```

```python
# Install necessary packages
!pip install bokeh pandas

# Import required libraries
import math
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool, LabelSet
from bokeh.io import output_notebook
from bokeh.palettes import Category10

# Sample data: Cities and their latitudes, longitudes
cities = pd.DataFrame({
    'city': ['New York', 'London', 'Tokyo', 'Sydney', 'Paris', 'Berlin'],
    'latitude': [40.7128, 51.5074, 35.6895, -33.8688, 48.8566, 52.52],
    'longitude': [-74.0060, -0.1278, 139.6917, 151.2093, 2.3522, 13.4050]
})

# Example connections (edges) between cities (represented by index of cities)
connections = [
    (0, 1), (1, 2), (2, 3), (3, 4), (4, 5),  # Connections forming a chain
    (0, 2), (1, 3), (2, 4), (3, 5)  # Some additional connections
]

# Function to convert latitude and longitude to Web Mercator (simplified)
def lat_lon_to_mercator(lat, lon):
    x = lon * 20037508.34 / 180  # Longitude to Web Mercator X
    y = math.log(math.tan((90 + lat) * math.pi / 360)) * 20037508.34 / math.pi  # Latitude to Web Mercator Y
    return x, y

# Convert lat/lon to Web Mercator coordinates for cities
cities['x'], cities['y'] = zip(*[lat_lon_to_mercator(lat, lon) for lat, lon in zip(cities['latitude'], cities['longitude'])])

# Manually assign colors from Category10 palette to each city
cities['color'] = Category10[6]

# Create a Bokeh data source for cities
city_source = ColumnDataSource(cities)

# Set up the Bokeh plot
p = figure(title="City Network Map", width=800, height=500)

# Add city markers with color assigned
p.circle(x='x', y='y', size=10, color='color', alpha=0.7, legend_label="Cities", source=city_source)

# Add labels for city names
labels = LabelSet(x='x', y='y', text='city', level='glyph', x_offset=5, y_offset=5,
                  source=city_source, text_font_size='10pt')

# Attach the labels to the circle glyphs by specifying the 'renderers' attribute
p.add_layout(labels)

# Plot the connections (edges) between cities
for connection in connections:
    city1 = cities.iloc[connection[0]]
    city2 = cities.iloc[connection[1]]
    p.line(x=[city1['x'], city2['x']], y=[city1['y'], city2['y']], line_width=2, color="green", alpha=0.5)

# Add hover tool for city information
hover = HoverTool()
hover.tooltips = [("City", "@city"), ("Latitude", "@latitude"),
                  ("Longitude", "@longitude")]
p.add_tools(hover)

# Show the plot
output_notebook()
show(p)
```

```python
# Install required libraries in Google Colab
!pip install requests beautifulsoup4

import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os

# Function to retrieve an image over HTTP
def retrieve_image(image_url, save_path):
    try:
        # Send an HTTP GET request to the image URL
        response = requests.get(image_url)
        if response.status_code == 200:
            # Save the image content to a file
            with open(save_path, 'wb') as file:
                file.write(response.content)
            print(f"Image saved as {save_path}")
        else:
            print(f"Failed to retrieve image: {image_url}")
    except Exception as e:
        print(f"Error retrieving image: {e}")

# Function to parse HTML and scrape content
def scrape_website(url):
    try:
        # Send an HTTP GET request to the website URL
        response = requests.get(url)
        if response.status_code == 200:
            # Parse the page using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            # Print the title of the webpage
            title = soup.title.string if soup.title else "No title found"
            print(f"Webpage title: {title}")

            # Extract and display all image URLs
            images = soup.find_all('img')
            if images:
                print(f"Found {len(images)} image(s) on the webpage.")
                for idx, img in enumerate(images):
                    img_url = img.get('src')
                    if img_url:
                        img_url = urljoin(url, img_url)  # Handle relative URLs
                        image_name = f"image_{idx + 1}.jpg"
                        save_path = f"/content/{image_name}"
                        retrieve_image(img_url, save_path)
            else:
                print("No images found on the webpage.")
        else:
            print("Failed to retrieve webpage")
    except Exception as e:
        print(f"Error scraping website: {e}")

# Example usage for scraping and image retrieval
webpage_url = 'https://www.wikipedia.org/'  # Wikipedia homepage for web scraping
scrape_website(webpage_url)
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Set up the Seaborn style and context
sns.set(style="whitegrid", palette="muted", font_scale=1.2)

# Create a pairplot for the iris dataset
pairplot = sns.pairplot(iris, hue="species", markers=["o", "s", "D"], plot_kws={'alpha': 0.6})

# Set titles and labels
pairplot.fig.suptitle('Pairplot of Iris Dataset', fontsize=18)
pairplot.fig.tight_layout()  # Ensures that titles and labels are not cut off
pairplot.fig.subplots_adjust(top=0.95)  # Adjusts the spacing to accommodate the title

# Show the plot
plt.show()
```
